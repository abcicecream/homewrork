```python
import platform
import matplotlib as mpl
from matplotlib import font_manager, rc
system = platform.system()
if system == "Windows":
    font_path = "C:/Windows/Fonts/malgun.ttf"
elif system == "Darwin":
    font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc"
elif system == "Linux":
    font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
else:
    font_path = None
if font_path:
    font = font_manager.FontProperties(fname=font_path).get_name()
    mpl.rcParams['font.family'] = font
else:
    print("폰트 경로를 확인해주세요.")
mpl.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import matplotlib.pyplot as plt

excel_data = pd.read_excel('notExercise.xlsx')
excel_data
```

```python
data_s = ['남자','여자']
data_a = ['10대','20대','30대','40대','50대','60대 이상']
data_e = ['중졸 이하','고졸 이하','대졸 이하','대학원 이상']

time = '운동을 할 충분한 시간이 없어서'
peo = '함께 운동을 할 사람이 없어서'
pla = '운동을 할 만한 장소가 없어서'
hat = '운동을 싫어해서'

s_time = excel_data['운동을 할 충분한 시간이 없어서'][1:3]
s_peo = excel_data['함께 운동을 할 사람이 없어서'][1:3]
s_pla = excel_data['운동을 할 만한 장소가 없어서'][1:3]
s_hat = excel_data['운동을 싫어해서'][1:3]

a_time = excel_data['운동을 할 충분한 시간이 없어서'][3:9]
a_peo = excel_data['함께 운동을 할 사람이 없어서'][3:9]
a_pla = excel_data['운동을 할 만한 장소가 없어서'][3:9]
a_hat = excel_data['운동을 싫어해서'][3:9]

e_time = excel_data['운동을 할 충분한 시간이 없어서'][9:13]
e_peo = excel_data['함께 운동을 할 사람이 없어서'][9:13]
e_pla = excel_data['운동을 할 만한 장소가 없어서'][9:13]
e_hat = excel_data['운동을 싫어해서'][9:13]
```

```python
# 성별

plt.figure(figsize=(5,5))
plt.pie(s_time, labels=data_s,autopct='%.1f%%')
plt.title(time)
plt.show()

plt.figure(figsize=(5,5))
plt.pie(s_peo, labels=data_s,autopct='%.1f%%')
plt.title(peo)
plt.show()

plt.figure(figsize=(5,5))
plt.pie(s_pla, labels=data_s,autopct='%.1f%%')
plt.title(pla)
plt.show()

plt.figure(figsize=(5,5))
plt.pie(s_hat, labels=data_s,autopct='%.1f%%')
plt.title(hat)
plt.show()
```

```python
# 연령별

plt.figure(figsize=(5,5))
plt.pie(a_time, labels=data_a,autopct='%.1f%%')
plt.title(time)

plt.figure(figsize=(5,5))
plt.pie(a_peo, labels=data_a,autopct='%.1f%%')
plt.title(peo)

plt.figure(figsize=(5,5))
plt.pie(a_pla, labels=data_a,autopct='%.1f%%')
plt.title(pla)

plt.figure(figsize=(5,5))
plt.pie(a_hat, labels=data_a,autopct='%.1f%%')
plt.title(hat)
plt.show()
```

```python
# 학력별

plt.figure(figsize=(5,5))
plt.pie(e_time, labels=data_e,autopct='%.1f%%')
plt.title(time)

plt.figure(figsize=(5,5))
plt.pie(e_peo, labels=data_e,autopct='%.1f%%')
plt.title(peo)

plt.figure(figsize=(5,5))
plt.pie(e_pla, labels=data_e,autopct='%.1f%%')
plt.title(pla)

plt.figure(figsize=(5,5))
plt.pie(e_hat, labels=data_e,autopct='%.1f%%')
plt.title(hat)
plt.show()
```
